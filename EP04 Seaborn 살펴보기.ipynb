```python
%matplotlib inline 
from matplotlib import pyplot as plt
import seaborn as sns

rc = {
    'axes.unicode_minus' : False,
    'font.family' : 'Malgun Gothic',
}

sns.set_style('whitegrid', rc)
sns.set_palette('bright')

```

```python
import pandas as pd
```

```python
df = pd.read_excel("data/sample-melon.xlsx").set_index('곡일련번호')
print(df.shape)
df.head()
```

```python
df['가수'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(30,8))
sns.countplot(data=df, x='가수')
plt.xticks(rotation=90)
```

```python
df.groupby('가수').size()
```

```python
df['가수'].value_counts().sort_values(ascending=False)[:10]
```

```python

with sns.color_palette('husl') as 팔레트:
    series = df.groupby('가수').size()
    series.plot(kind='bar', figsize=(30,8), color=팔레트)
```
